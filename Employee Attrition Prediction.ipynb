```python
# Exclude 'Attrition' column from outlier detection
numeric_cols = green.select_dtypes(include=['number']).columns.drop('Attrition')

Q1 = green[numeric_cols].quantile(0.25)
Q3 = green[numeric_cols].quantile(0.75)

#Calculate IQR - Inter Quartile Range 
IQR = Q3 - Q1

#Determine the outlier boundaries 
lower_bound = Q1 - 1.5 * IQR 
upper_bound = Q3 + 1.5 * IQR

#Create a mask for data within the IQR boundaries 
mask = ~((green[numeric_cols] < lower_bound) | (green[numeric_cols] > upper_bound)).any(axis=1)

#Remove outliers 
cleaned_data = green[mask]
```

```python
x = cleaned_data.drop('Attrition', axis=1)
x
```

```python
y = cleaned_data['Attrition']
y
```

```python
# SMOTE - Synthetic Minority Oversampling Techniques
# Smote is a method used in machine learningto address class imbalance in datasets. 
from imblearn.over_sampling import SMOTE

smote = SMOTE()
x_resampled, y_resampled = smote.fit_resample(x,y)
```

```python
# Checking
x_resampled.shape
```

```python
x.shape
```

```python
y_resampled.shape
```

```python
y.shape
```

```python
# Assigning variables for training and testing
x_train, x_test, y_train, y_test = train_test_split(x_resampled, y_resampled, test_size=0.3, random_state=42)
```

```python
# Checking shape
x_resampled.shape, x_train.shape, x_test.shape, y_resampled.shape, y_test.shape, y_train.shape
```

```python
# Model assigning
model = LogisticRegression()
```

```python
# Model fitting
model.fit(x_train,y_train)
```

```python
# Prediction
train_data_predict = model.predict(x_train)
train_data_accuracy = accuracy_score(train_data_predict,y_train)
print('Accuracy of training data',train_data_accuracy)
```

```python
test_data_predict = model.predict(x_test)
test_data_accuracy = accuracy_score(test_data_predict,y_test)
print('Accuracy of testing data',test_data_accuracy)
```

```python
# Confusion matrix
cm = confusion_matrix(test_data_predict, y_test)
cm
```

```python
sns.heatmap(cm, annot = True, annot_kws = {'size':20}, fmt = '.2f', cmap = 'viridis', square = True,
           xticklabels = ['True Positives','False Negatives'],
           yticklabels = ['False Positives','True Negatives'])
plt.xlabel('Predicted Data')
plt.ylabel('Actual Data')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Classification Report
print(classification_report(test_data_predict,y_test))
```
